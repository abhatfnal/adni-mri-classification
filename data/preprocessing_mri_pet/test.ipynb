```python
import pandas as pd 
import os
import zipfile
```

```python
# Path of PET scans
pet_zip_path = "/project/aereditato/abhat/ADNI/Original/ADNI_all_PET_Original_2.zip"
```

```python
# Read pairs csv
df = pd.read_csv("mri_pet_pairs.csv")
print(len(df))
df
```

```python
# Read files in pet archive
zip_ref = zipfile.ZipFile(pet_zip_path, 'r')

image_ids = []
for path in zip_ref.namelist():
    image_id = int(path.split("/")[4][1:])
    image_ids.append(image_id)
    
image_ids = list(set(image_ids))
```

```python
image_ids
```

```python
# Count number of pet scans that are missing from the zip file
count = 0

for id in df["pet_id"]:
    
    if int(id) not in image_ids:
        count += 1
        
print(count)
```
